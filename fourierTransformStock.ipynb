```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# df = "Put your stock data here. Should be a pandas Dataframe with Column 'High','Low','Open','Close'"
df = pd.read_csv("000040.csv")
f = np.array((df.High+df.Low+df.Open+df.Close)/4)
t = np.arange(0,len(f),1)
a = sum(f)

plt.plot(t,f)
plt.title('Stock Value')
plt.show()

##define min period and max period.
##default periods are minPeriod:2, maxPeriod:100, seperateBy:1 (unit:days)
## !! Do not set 1 as minPeriod. Since data is discrete, it will return the sum of all stock prices. !!

periods = np.arange(2,100,0.5)
sins = [np.sum(np.sin(2*np.pi*t/period)*f)/(2*len(t)) for period in periods]
coss = [np.sum(np.cos(2*np.pi*t/period)*f)/(2*len(t)) for period in periods]

##X: period of waves, Y: contribution of the wave whose period is x.
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(periods,sins,label='sin',color='purple')
plt.plot(periods,coss,label='cos',color='skyblue')
plt.legend()

plt.subplot(122)
##Plot absolute value of cos+sin*j
plt.plot(periods,np.sqrt(np.power(sins,2)+np.power(coss,2)),label="abs",color='grey')
plt.legend()
plt.show()
```
